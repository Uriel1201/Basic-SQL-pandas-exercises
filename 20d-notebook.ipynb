```python
import pandas as pd 
import numpy  as np

data = {'visitors'   : [10,109,150,99,145,1455,199,188],
        'event_date' : ['01-JAN-20',
                        '04-JAN-20',
                        '05-JAN-20',
                        '06-JAN-20',
                        '07-JAN-20',
                        '08-JAN-20',
                        '11-JAN-20',
                        '12-JAN-20']
        }
                    

attendance = pd.DataFrame(data)
```

```python
attendance 
```

```python
attendance['event_date'] = pd.to_datetime(attendance['event_date'])
attendance
```

```python
df = attendance.copy()
df.rename(columns = {'event_date':'event_date2','visitors':'visitors2'}, inplace = True)
df.drop(index = 0,  inplace = True)
df.reset_index(inplace = True)
df.drop(axis = 1, columns = 'index', inplace = True)
df1 = df.copy()
df1.rename(columns = {'event_date2':'event_date3','visitors2':'visitors3'}, inplace = True)
df1.drop(index = 0, inplace =  True)
df1.reset_index(inplace = True)
df1.drop(axis = 1, columns = 'index', inplace = True)
```
