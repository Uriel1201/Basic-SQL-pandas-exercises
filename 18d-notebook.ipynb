```python
import pandas as pd 
import numpy  as np

data = {'id'        : [1,2,3,4,5,6,7,8,9],
        'city'      : ['Asheville','Burlington','Chapel Hill','Davidson','Elizabeth City','Fargo',
                       'Grand Forks','Hettinger','Inkster'],
        'states'    : ['North Carolina','North Carolina','North Carolina','North Carolina','North Carolina',
                       'North Dakota','North Dakota','North Dakota','North Dakota'],
        'latitude'  : [35.6,36.1,35.9,35.5,36.3,46.9,47.9,46,48.2],
        'longitude' : [82.6,79.4,79.1,80.8,76.3,96.8,97,102.6,97.6]
        }

stations = pd.DataFrame(data)
```

```python
stations 
```

```python
cities1 = stations.copy()
cities1.rename(columns = {'city':'city1','latitude':'lat1','longitude':'lon1'}, inplace = True)
cities2 = stations.copy()
cities2.rename(columns = {'city':'city2','latitude':'lat2','longitude':'lon2'}, inplace = True)
df = pd.merge(cities1,cities2, on = 'states', how = 'inner')
df
```

```python
df = df[df.id_x < df.id_y][['states','city1','lat1','lon1','city2','lat2','lon2']]
```

```python
df
```

```python
df['distance'] = np.sqrt(((df['lat2']-df['lat1'])*(df['lat2']-df['lat1']) + (df['lon2']-df['lon1'])*(df['lon2']-df['lon1'])))
df['distance'] = df['distance'].round(2)       
```

```python
df
```

```python
df['ranks'] = df.groupby('states')['distance'].rank(method = 'first',ascending = False)
df[df.ranks == 1][['states','city1','city2','distance']]
```
