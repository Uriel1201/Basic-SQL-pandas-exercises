```python
import pandas as pd 
import numpy  as np

data1 = {'order_id'    : [1,1,1,2,2,2,3],
        'customer_id' : [1,1,1,2,2,2,1],
        'product_id'  : [1,2,3,1,2,4,5]
        }

data2 = {'id'   : [1,2,3,4,5],
         'name' : ['A','B','C','D','E']
        }

orders   = pd.DataFrame(data1)
products = pd.DataFrame(data2)
```

```python
orders
```

```python
products 
```

```python
orders1 = orders[['order_id','product_id']].copy()
orders2 = orders[['order_id','product_id']].copy()
```

```python
df = pd.merge(orders1, orders2, on = 'order_id', how = 'inner')
```

```python
df = df[df.product_id_x < df.product_id_y]
```

```python
df
```

```python
df1 = products.copy()
df1.rename(columns = {'id':'product_id_x'}, inplace = True)
df = pd.merge(df, df1, on = 'product_id_x', how = 'left')
```

```python
df.rename(columns = {'name':'name_x'}, inplace  = True)
df
```

```python
df1.rename(columns = {'product_id_x':'product_id_y'}, inplace = True)
df = pd.merge(df, df1, on = 'product_id_y', how = 'left')
```

```python
df.rename(columns = {'name':'name_y'}, inplace = True)
```

```python
df
```

```python
df['pairs'] = df['name_x'].str.cat(df['name_y'])
```

```python
df = df[['order_id','pairs']]
df
```

```python
df = pd.DataFrame(df.groupby('pairs').size())
```

```python
df.rename(columns = {0:'frequency'}, inplace = True)
df.sort_values(by = 'frequency', ascending = False)
```
