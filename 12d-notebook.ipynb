```python
import pandas as pd 
import numpy  as np 

data1 = {'hacker_id' : [1,2,3,4],
         'name'      : ['John','Jane','Joe','Jim']
        }

data2 = {'sub_id'       : [101,102,103,104,105,106,107,108,109],
         'hacker_id'    : [1,1,2,2,2,3,3,3,4],
         'challenge_id' : [1,1,1,1,2,1,2,2,1],
         'score'        : [10,12,11,9,13,9,12,15,0]
        }

hackers     = pd.DataFrame(data1)
submissions = pd.DataFrame(data2)
```

```python
hackers 
```

```python
submissions
```

```python
submissions['ranks'] = submissions.groupby(['hacker_id','challenge_id'])['score'].rank(method = 'min', ascending = False)
submissions 
```

```python
merges     = pd.merge(hackers, submissions, on = 'hacker_id', how = 'left')
max_score  = merges[merges.ranks == 1][['hacker_id','name','score']].copy()
max_score.reset_index(inplace = True)
max_score.drop(['index'],axis = 1, inplace = True)
max_score 
```

```python
total_scores = pd.DataFrame(max_score.groupby(['hacker_id','name'])['score'].sum())
total_scores.reset_index(inplace = True)
total_scores.rename(columns = {'score':'total_score'}, inplace = True)
total_scores[total_scores.total_score > 0].sort_values(by = ['total_score','hacker_id'], ascending = [False,True])
```
