```python
import pandas as pd 
import numpy  as np 

data1 = {'id'   : [1,2,3,4],
         'name' : ['Daniel','Diana','Elizabeth','John']
        }

data2 = {'order_id'     : [1,2,3,4,5,6,7,8,9],
         'customer_id'  : [1,1,2,2,2,3,3,3,3],
         'product_name' : ['A','B','A','B','C','A','A','B','D'],
        }

customers = pd.DataFrame(data1)
orders    = pd.DataFrame(data2)
```

```python
customers 
```

```python
orders
```

```python
A = pd.DataFrame(orders[orders.product_name == 'A'].drop_duplicates('customer_id')['customer_id'])
```

```python
B = pd.DataFrame(orders[orders.product_name == 'B'].drop_duplicates('customer_id')['customer_id'])
```

```python
C = pd.DataFrame(orders[orders.product_name == 'C'].drop_duplicates('customer_id')['customer_id'])
```

```python
AB = pd.merge(A, B, on = 'customer_id', how = 'inner')
```

```python
AB
```

```python
noC = AB[~AB.customer_id.isin(C['customer_id'])]
noC
```

```python
customer2 = customers.copy()
customer2.rename(columns = {'id':'customer_id'}, inplace = True)
noC = pd.merge(noC, customer2, on = 'customer_id', how = 'left')
noC
```
