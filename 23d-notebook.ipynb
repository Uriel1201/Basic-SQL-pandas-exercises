```python
import pandas as pd 
import numpy  as np

data = {'months' : [1,2,3,4,5,6,7,8,9],
        'salary' : [2000,3000,5000,4000,2000,1000,2000,4000,5000]
       }
salary = pd.DataFrame(data)
```

```python
salary
```

```python
df1 = salary.copy()
df2 = salary.copy()
```

```python
df1.drop(index = 0, inplace = True)
df2.drop(index = [0,1], inplace = True)
```

```python
df1
```

```python
df2
```

```python
df1.reset_index(inplace = True)
df2.reset_index(inplace = True)
```

```python
df1.rename(columns = {'salary':'salary2'}, inplace = True)
df2.rename(columns = {'salary':'salary3'}, inplace = True)
df1 = df1['salary2'].copy()
df2 = df2['salary3'].copy()
```

```python
df1 = pd.DataFrame(df1)
df2 = pd.DataFrame(df2)
```

```python
df1
```

```python
df2
```

```python
df = salary.join(df1)
df = df.join(df2)
```

```python
df
```

```python
df['cummulative_3months']= df.apply(lambda x: x['salary'] + x['salary2'] + x['salary3'], axis = 1)
```
