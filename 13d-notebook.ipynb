```python
import pandas as pd 
import numpy  as np 

data = {'id'    : [1,2,3,4,5,6],
        'score' : [3.5,3.65,4.0,3.85,4.0,3.65]
        }

scores = pd.DataFrame(data)
```

```python
scores
```

```python
s = scores.drop_duplicates('score').sort_values(by = 'score', ascending = False).copy()
s.reset_index(inplace = True)
s = s['score'].copy()
s
```

```python
gauge = []
rows  = [v for v in s]
for i in rows:
    n = pd.Index(s).get_loc(i) + 1
    gauge.append(n)
```

```python
gauge = pd.DataFrame(gauge)
s     = pd.DataFrame(s)
ranks = pd.concat([s, gauge], axis = 1)
ranks.rename(columns = {0:'rank_score'}, inplace = True)
```

```python
ranks
```

```python
ranks = pd.merge(scores, ranks, on = 'score')
ranks[['id','rank_score']].sort_values(by = 'rank_score')
```
