```python
import pandas as pd 
import numpy as  np 

data = {#'user_id' :[1,1,2,1,1,2,3,3,4],
        'item'  :['apple','apple','pear',
                    'pear','pear','pear',
                    'pear','orange'],
        'dates'   :['01-JAN-20',
                    '01-JAN-20',
                    '01-JAN-20',
                    '01-JAN-20',
                    '02-JAN-20',
                    '02-JAN-20',
                    '02-JAN-20',
                    '02-JAN-20']
        }

items = pd.DataFrame(data)
items
```

```python
item_counts = items.groupby(['dates','item']).agg({'item':np.size})
item_counts.rename(columns = {'item':'count'}, inplace = True)
item_counts.reset_index(inplace = True)
item_counts 
```

```python
item_counts['ranks'] = item_counts.groupby('dates')['count'].rank(method = 'min', ascending = False)
item_counts 
```

```python
item_counts[item_counts.ranks == 1.0][['dates','item']]
```
