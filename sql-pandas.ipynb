```python
import pandas as pd 
import numpy as np 

data = {'sender'          :[5,1,2,2,3,3,1],
        'receiver'        :[2,3,1,3,1,2,4],
        'amount'          :[10,15,20,25,20,15,5],
        'transaction_date':['12-feb-20',
                            '13-feb-20',
                            '13-feb-20',
                            '14-feb-20',
                            '15-feb-20',
                            '15-feb-20',
                            '16-feb-20']
        }

transactions = pd.DataFrame(data)
transactions
```

```python
debited  = transactions[['sender','amount']].copy()
credited = transactions[['receiver','amount']].copy()

debited.rename(columns  = {'sender'  :'user_id','amount'  :'debited'}, inplace = True)
credited.rename(columns = {'receiver':'user_id','amount'  :'credited'},inplace = True)
```

```python
debited
```

```python
credited
```

```python
all_debited  = debited.groupby('user_id').sum().copy()
all_credited = credited.groupby('user_id').sum().copy()
```

```python
all_debited
```

```python
all_credited
```

```python
df = pd.merge(all_debited, all_credited, on = 'user_id',how = 'outer')
```

```python
df
```

```python
df['debited']  = df['debited'].replace(np.nan, 0)
df['credited'] = df['credited'].replace(np.nan,0)
df
```

```python
df['net_change'] = df['credited'] - df['debited']
```
