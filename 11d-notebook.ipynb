```python
import pandas as pd 
import numpy  as np

data1 = {'student_id'    : [1,2,3,1,2,3,1,2,3,4],
        'school_date'    : ['03-APR-20',
                            '03-APR-20',
                            '03-APR-20',
                            '04-APR-20',
                            '04-APR-20',
                            '04-APR-20',
                            '05-APR-20',
                            '05-APR-20',
                            '05-APR-20',
                            '05-APR-20'],
         'attendance'    : ['F','T','T','T','T','T','F','T','T','T']
        }

data2 = {'student_id'    : [1,2,3,4],
         'school_id'     : [2,1,1,2],
         'grade'         : [5,4,3,4],
         'date_birth'    : ['03-APR-12',
                            '04-APR-13',
                            '05-APR-14',
                            '03-APR-13']
                        # '15-OCT-20',
                        # '29-OCT-20',
                        # '31-OCT-20']
        }

attendance  = pd.DataFrame(data1)
students    = pd.DataFrame(data2)
```

```python
attendance['school_date'] = pd.to_datetime(attendance['school_date'])
students['date_birth']    = pd.to_datetime(students['date_birth'])
```

```python
merge = pd.merge(students, attendance, on = 'student_id', how = 'inner')
filtered            = np.where((merge['school_date'].dt.day == merge['date_birth'].dt.day) & (merge['school_date'].dt.month == merge['date_birth'].dt.month))
birthday_attendance = merge.loc[filtered][['student_id','attendance']].copy()
birthday_attendance.rename(columns = {'attendance':'birthday_attendance'}, inplace = True)
birthday_attendance
```

```python
i = (np.where(birthday_attendance['birthday_attendance'] == 'T', 1, 0).sum() / len(birthday_attendance.index)).round(2)
d = {'birthday_attendance_rate' : i
  }
birthday_attendance_rate = pd.DataFrame(data = d, index = [0])
birthday_attendance_rate
```
