```python
import pandas as pd 
import numpy  as np 

data1 = {'team_id'   : [1,2,3,4,5,6],
         'team_name' : ['New York','Atlanta','Chicago','Toronto','Los Angeles','Seattle']
        }

data2 = {'match_id'    : [1,2,3,4,5,6],
         'host_team'   : [1,2,3,4,5,6],
         'guest_team'  : [2,3,4,5,6,1],
         'host_goals'  : [3,2,4,1,2,1],
         'guest_goals' : [0,4,3,1,1,2]
        }

teams   = pd.DataFrame(data1)
matches = pd.DataFrame(data2)
```

```python
teams
```

```python
matches 
```

```python
matches['host_score']  = np.where(matches['host_goals'] > matches['guest_goals'], 3, np.where(matches['host_goals'] < matches['guest_goals'], 0, 1))
matches['guest_score'] = np.where(matches['host_goals'] < matches['guest_goals'], 3, np.where(matches['host_goals'] > matches['guest_goals'], 0, 1))
```

```python
matches 
```

```python
hosts  = matches[['host_team','host_score']].copy()
guests = matches[['guest_team','guest_score']].copy()
hosts.rename(columns  = {'host_team':'team_id'}, inplace  = True)
guests.rename(columns = {'guest_team':'team_id'}, inplace = True)
```

```python
hosts
```

```python
guests
```

```python
hosts  = hosts.groupby('team_id').sum()
guests = guests.groupby('team_id').sum()
```

```python
hosts
```

```python
guests 
```

```python
guests['host_score'] = hosts
```

```python
guests
```

```python
guests['total_score'] = guests['host_score'] + guests['guest_score']
```

```python
guests.reset_index(inplace = True)
```

```python
guests
```

```python
scores = pd.merge(guests, teams, on = 'team_id')
```

```python
scores = scores[['team_name','total_score']].sort_values(by = ['total_score','team_name'], ascending = [False,True])
```
