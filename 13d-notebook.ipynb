```python
import pandas as pd 
import numpy  as np 

data = {'id'    : [1,2,3,4,5,6],
        'score' : [3.5,3.65,4.0,3.85,4.0,3.65]
        }

scores = pd.DataFrame(data)
```

```python
scores
```

```python
distinct = scores.drop_duplicates('score')['score'].copy()
gauge    = []
rows     = [v for v in distinct]
x        = rows
y        = rows

for i in x:
    for j in y:
        if i <= j:
            gauge.append(i)
```

```python
gauge = pd.DataFrame(gauge)
gauge.rename(columns = {0:'score'}, inplace = True)
ranks = gauge.groupby('score').agg({'score':np.size})
ranks.rename(columns = {'score':'rank_score'}, inplace = True)
ranks.reset_index(inplace = True)
ranks
```

```python
ranks = pd.merge(scores, ranks, on = 'score')
ranks.sort_values(by = 'rank_score')
```
